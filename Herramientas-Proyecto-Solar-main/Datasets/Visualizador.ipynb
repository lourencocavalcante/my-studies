```python
import os
import h5py
import matplotlib.pyplot as plt
```

```python
# Lista de datasets

lista_datasets = os.listdir("Datasets/")
print("Lista de datasets disponibles:\n")
for i,archivo in zip(range(len(lista_datasets)),lista_datasets):
    print(i,archivo)
```

```python
# Seleccionamos dataset
index_dataset = 0 

# Abrimos dataset y vemos contenido
nombre_dataset = "Datasets/" + lista_datasets[index_dataset]
print("Contenido del archivo:\n")
with h5py.File(nombre_dataset,"r") as dataset: print(dataset.keys())
```

```python
print("Extraemos los datos:\n")

with h5py.File(nombre_dataset,"r") as dataset:
    banda_13 = dataset["13"][()]
    banda_04 = dataset["4"][()]
    GHI = dataset["GHI"][()]
    Clear = dataset["Clearsky GHI"][()]
    
print(banda_13.shape)
print(banda_04.shape)
print(Clear.shape)
print(GHI.shape)
```

```python
# Visualizamos

index = 45

print(f"GHI {GHI[index]} W/m², Clearsky {Clear[index]} W/m²")
fig, ax = plt.subplots(1,2,figsize=(10,5))
ax[0].imshow(banda_04[index])
ax[1].imshow(banda_13[index])
```
